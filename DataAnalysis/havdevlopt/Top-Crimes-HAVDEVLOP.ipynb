```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')


#load csv
overall = pd.read_csv("overall-crime.csv", header=None,)
overall.columns = ['HADEVELOPT', 'COUNT']
overall=overall[overall['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll

bronx = pd.read_csv("bronx-crime.csv", header=None,)
bronx.columns = ['HADEVELOPT', 'COUNT']
bronx=bronx[bronx['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll

brooklyn = pd.read_csv("brooklyn-crime.csv", header=None,)
brooklyn.columns = ['HADEVELOPT', 'COUNT']
brooklyn=brooklyn[brooklyn['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll

manhattan = pd.read_csv("manhattan-crime.csv", header=None,)
manhattan.columns = ['HADEVELOPT', 'COUNT']
manhattan=manhattan[manhattan['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll

staten = pd.read_csv("staten-crime.csv", header=None,)
staten.columns = ['HADEVELOPT', 'COUNT']
staten=staten[staten['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll

queens = pd.read_csv("queens-crime.csv", header=None,)
queens.columns = ['HADEVELOPT', 'COUNT']
queens=queens[queens['HADEVELOPT'].notnull()] #pandas interprets NA by default as nulll
```

```python
#kycd & ofnc top 10 
overall=overall.sort_values(by='COUNT', ascending=False)
overall = overall.head(10)
overall.to_csv('top-overall.csv')
```

```python
#staten
staten = staten.sort_values(by='COUNT', ascending=False)
staten=staten.head(10)
staten.to_csv('top-staten.csv')
```

```python
#bronx
bronx = bronx.sort_values(by='COUNT', ascending=False)
bronx=bronx.head(10)
bronx.to_csv('top-bronx.csv')
```

```python
#brooklyn
brooklyn = brooklyn.sort_values(by='COUNT', ascending=False)
brooklyn=brooklyn.head(10)
brooklyn.to_csv('top-brooklyn.csv')
```

```python
#manhttan
manhattan=manhattan.sort_values(by='COUNT', ascending=False)
manhattan=manhattan.head(10)
manhattan.to_csv('top-manhattan.csv')
```

```python
#queens
queens=queens.sort_values(by='COUNT', ascending=False)
queens=queens.head(10)
queens.to_csv('top-queens.csv')
```

```python
#overall
#pivot_overall = overall.pivot(index='OFNS_DESC', columns='KY_CD', values='COUNT')
pivot_overall = overall.set_index('HADEVELOPT')
pivot_overall.plot(kind='bar',stacked=True,)
#plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8))
plt.title("Top 10 Areas in NYC with Highest Number of Crimes")
plt.xlabel('HADEVELOPT')
plt.ylabel('COUNT')
plt.savefig("overall-crime-hav.png",bbox_inches='tight')
```
