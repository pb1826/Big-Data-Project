```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')


#load csv
prime22 = pd.read_csv("precinct-crime-22.csv", header=None,)
prime22.columns = ['KY_CD','OFSC_DESC', 'COUNT']

#ascending
prime22asc=prime22.sort_values(by='COUNT')
prime22asc = prime22asc.head(10)
prime22asc.to_csv('bottom-precint22.csv')

#descending
prime22dsc=prime22.sort_values(by='COUNT',ascending=False)
prime22dsc = prime22dsc.head(10)
prime22dsc.to_csv('top-precint22.csv')
```

```python
#plot ascending
pivot_asc = prime22asc.pivot(index='OFSC_DESC', columns='KY_CD', values='COUNT')
pivot_asc.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.title("Lowest 10 crimes in Central Prak Precinct")
plt.xlabel('OFSC_DESC')
plt.ylabel('COUNT')
plt.savefig("22-crime-precinct-low.png",bbox_inches='tight')
```

```python
#plot descending
pivot_dsc = prime22dsc.pivot(index='OFSC_DESC', columns='KY_CD', values='COUNT')
pivot_dsc.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.title("Highest 10 crimes in Central Prak Precinct")
plt.xlabel('OFSC_DESC')
plt.ylabel('COUNT')
plt.savefig("22-crime-precinct-high.png",bbox_inches='tight')
```
