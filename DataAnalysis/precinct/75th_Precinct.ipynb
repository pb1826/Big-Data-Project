```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')


#load csv
prime75_top = pd.read_csv("precinct-crime-75.csv", header=None,)
prime75_top.columns = ['KY_CD','OFSC_DESC', 'COUNT']

#load csv
prime75_bottom = pd.read_csv("precinct-crime-75.csv", header=None,)
prime75_bottom.columns = ['KY_CD','OFSC_DESC', 'COUNT']

#descending
prime75dsc=prime75_top.sort_values(by='COUNT',ascending=False)
prime75dsc = prime75dsc.head(10)
prime75dsc.to_csv('top-precint75.csv')

#ascending
prime75asc=prime75_bottom.sort_values(by='COUNT')
prime75asc = prime75asc.head(10)
prime75asc.to_csv('bottom-precint75.csv')
```

```python
#plot ascending
pivot_asc = prime75asc.pivot(index='OFSC_DESC', columns='KY_CD', values='COUNT')
pivot_asc.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.title("Lowest 10 crimes in 75th Precinct")
plt.xlabel('OFSC_DESC')
plt.ylabel('COUNT')
plt.savefig("75-crime-precinct-low.png",bbox_inches='tight')
```

```python
#plot descending
pivot_dsc = prime75dsc.pivot(index='OFSC_DESC', columns='KY_CD', values='COUNT')
pivot_dsc.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.title("Highest 10 crimes in 75th Precinct")
plt.xlabel('OFSC_DESC')
plt.ylabel('COUNT')
plt.savefig("75-crime-precinct-high.png",bbox_inches='tight')
```
