```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')


#load csv
overall = pd.read_csv("overall-crime.csv", header=None,)
overall.columns = ['PD_CD', 'PD_DESC', 'COUNT']
bronx = pd.read_csv("bronx-crime.csv", header=None,)
bronx.columns = ['PD_CD', 'PD_DESC', 'COUNT']

brooklyn = pd.read_csv("brooklyn-crime.csv", header=None,)
brooklyn.columns = ['PD_CD', 'PD_DESC', 'COUNT']

manhattan = pd.read_csv("manhattan-crime.csv", header=None,)
manhattan.columns = ['PD_CD', 'PD_DESC', 'COUNT']

staten = pd.read_csv("staten-crime.csv", header=None,)
staten.columns = ['PD_CD', 'PD_DESC', 'COUNT']

queens = pd.read_csv("queens-crime.csv", header=None,)
queens.columns = ['PD_CD', 'PD_DESC', 'COUNT']
```

```python
#kycd & ofnc top 10 
overall=overall.sort_values(by='COUNT')
overall = overall.head(10)
overall.to_csv('bottom-overall.csv')
```

```python
#staten
staten = staten.sort_values(by='COUNT')
staten=staten.head(10)
staten.to_csv('bottom-staten.csv')
```

```python
#bronx
bronx = bronx.sort_values(by='COUNT')
bronx=bronx.head(10)
bronx.to_csv('bottom-bronx.csv')
```

```python
#brooklyn
brooklyn = brooklyn.sort_values(by='COUNT')
brooklyn=brooklyn.head(10)
brooklyn.to_csv('bottom-brooklyn.csv')
```

```python
#manhttan
manhattan=manhattan.sort_values(by='COUNT')
manhattan=manhattan.head(10)
manhattan.to_csv('bottom-manhattan.csv')
```

```python
#queens
queens=queens.sort_values(by='COUNT')
queens=queens.head(10)
queens.to_csv('bottom-queens.csv')
```

```python
#overall
pivot_overall = overall.pivot(index='PD_DESC', columns='PD_CD', values='COUNT')
pivot_overall.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='PD_CD')
plt.title("Least 10 Crimes in NYC by PD")
plt.xlabel('PD_DESC')
plt.ylabel('COUNT')
plt.savefig("bottom-overall-crime.png",bbox_inches='tight')
```

```python
#bronx
pivot_bronx = bronx.pivot(index='PD_DESC', columns='PD_CD', values='COUNT')
pivot_bronx.plot(kind='bar',stacked=True,)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='PD_CD')
plt.title("Least 10 Crimes in Bronx by PD")
plt.xlabel('PD_DESC')
plt.ylabel('COUNT')
plt.savefig("bottom-bronx-crime.png",bbox_inches='tight')
```
