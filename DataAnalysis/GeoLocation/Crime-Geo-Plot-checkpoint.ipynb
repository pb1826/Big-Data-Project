```python
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import os
import pandas as pd
import seaborn as sns
import time
import bokeh.io, bokeh.plotting, bokeh.models
from bokeh.models import DatetimeTickFormatter
bokeh.io.output_notebook()
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, DataRange1d, PanTool, WheelZoomTool, BoxSelectTool
)
from bokeh.resources import CDN
from bokeh.embed import components, autoload_static, autoload_server
import folium


%matplotlib inline
```

```python
bronx = pd.read_csv("locationplot/bronx-assault-all.csv", names = ["lat","long"])
bronx["lat"] = bronx["lat"].str.replace("(","")
bronx["long"] = bronx["long"].str.replace(")","")
bronx['lat'] = bronx['lat'].astype(float)
bronx['long'] = bronx['long'].astype(float)
# get the coordinates with the most incidents of assault in bronx
new =  pd.core.frame.DataFrame({'count' : bronx.groupby(["lat", "long"] ).size()}).reset_index().sort_values('count',ascending = False)
new['index'] = new.index
new = new.head(50)
map = folium.Map(location=[40.8448, -73.8648], zoom_start=13)
new.apply(lambda row:folium.Marker(location=[row["lat"], row["long"]]).add_to(map),axis = 1)

map
```
