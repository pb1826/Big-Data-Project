```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import *
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')

#time limits
max=datetime.strptime('11:59:59', "%H:%M:%S").strftime('%H:%M:%S')
min=datetime.strptime('00:00:00', "%H:%M:%S").strftime('%H:%M:%S')

#load csv
overall = pd.read_csv("overall-crime.csv", header=None,)
overall.columns = ['TIME','OFNS_DESC','COUNT']
overall=overall[(min<=overall['TIME']) & (overall['TIME']<max)] #pandas interprets NA by default as nulll

bronx = pd.read_csv("bronx-crime.csv", header=None,)
bronx.columns = ['TIME','OFNS_DESC','COUNT']
bronx=bronx[(min<=bronx['TIME']) & (bronx['TIME']<max)] #pandas interprets NA by default as nulll

brooklyn = pd.read_csv("brooklyn-crime.csv", header=None,)
brooklyn.columns = ['TIME','OFNS_DESC','COUNT']
brooklyn=brooklyn[(min<=brooklyn['TIME']) & (brooklyn['TIME']<max)] #pandas interprets NA by default as nulll

manhattan = pd.read_csv("manhattan-crime.csv", header=None,)
manhattan.columns = ['TIME','OFNS_DESC','COUNT']
manhattan=manhattan[(min<=manhattan['TIME']) & (manhattan['TIME']<max)] #pandas interprets NA by default as nulll

staten = pd.read_csv("staten-crime.csv", header=None,)
staten.columns = ['TIME','OFNS_DESC','COUNT']
staten=staten[(min<=staten['TIME']) & (staten['TIME']<max)] #pandas interprets NA by default as nulll

queens = pd.read_csv("queens-crime.csv", header=None,)
queens.columns = ['TIME','OFNS_DESC','COUNT']
queens=queens[(min<=queens['TIME']) & (queens['TIME']<max)] #pandas interprets NA by default as nulll
```

```python
#kycd & ofnc bottom 10 
overall=overall.sort_values(by='COUNT')
overall = overall.head(10)
overall.to_csv('bottom-overall-am.csv')

#staten
staten = staten.sort_values(by='COUNT')
staten=staten.head(10)
staten.to_csv('bottom-staten-am.csv')

#bronx
bronx = bronx.sort_values(by='COUNT')
bronx=bronx.head(10)
bronx.to_csv('bottom-bronx-am.csv')

#brooklyn
brooklyn = brooklyn.sort_values(by='COUNT')
brooklyn=brooklyn.head(10)
brooklyn.to_csv('bottom-brooklyn-am.csv')

#manhttan
manhattan=manhattan.sort_values(by='COUNT')
manhattan=manhattan.head(10)
manhattan.to_csv('bottom-manhattan-am.csv')

#queens
queens=queens.sort_values(by='COUNT')
queens=queens.head(10)
queens.to_csv('bottom-queens-am.csv')
```

```python
#overall
pivot_overall = overall.pivot(index='TIME', columns='OFNS_DESC', values='COUNT')
#pivot_overall = overall.set_index('HADEVELOPT')
pivot_overall.plot(kind='bar',stacked=True,)
#plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='OFNS_DESC')
plt.legend(loc='upper right', bbox_to_anchor=(1.8, 0.8), title='OFNS_DESC')
plt.title("Times at which Crimes were Lowest (AM)")
plt.xlabel('TIME')
plt.ylabel('COUNT')
plt.savefig("overall-crime-time-low-am.png",bbox_inches='tight')
```
