```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')

#load csv
overall = pd.read_csv("overall-crime.csv", header=None,)
overall.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
overall=overall[overall['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll

bronx = pd.read_csv("bronx-crime.csv", header=None,)
bronx.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
bronx=bronx[bronx['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll

brooklyn = pd.read_csv("brooklyn-crime.csv", header=None,)
brooklyn.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
brooklyn=brooklyn[brooklyn['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll

manhattan = pd.read_csv("manhattan-crime.csv", header=None,)
manhattan.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
manhattan=manhattan[manhattan['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll

staten = pd.read_csv("staten-crime.csv", header=None,)
staten.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
staten=staten[staten['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll

queens = pd.read_csv("queens-crime.csv", header=None,)
queens.columns = ['OFNS_DESC', 'PARKS_NM','COUNT']
queens=queens[queens['PARKS_NM'].notnull()] #pandas interprets NA by default as nulll
```

```python
#kycd & ofnc bottom 10 
overall=overall.sort_values(by='COUNT')
overall = overall.head(10)
overall.to_csv('bottom-overall.csv')
#staten
staten = staten.sort_values(by='COUNT')
staten=staten.head(10)
staten.to_csv('bottom-staten.csv')
#bronx
bronx = bronx.sort_values(by='COUNT')
bronx=bronx.head(10)
bronx.to_csv('bottom-bronx.csv')
#brooklyn
brooklyn = brooklyn.sort_values(by='COUNT')
brooklyn=brooklyn.head(10)
brooklyn.to_csv('bottom-brooklyn.csv')
#manhttan
manhattan=manhattan.sort_values(by='COUNT')
manhattan=manhattan.head(10)
manhattan.to_csv('bottom-manhattan.csv')
#queens
queens=queens.sort_values(by='COUNT')
queens=queens.head(10)
queens.to_csv('bottom-queens.csv')
```

```python
#overall
pivot_overall = overall.pivot(index='PARKS_NM', columns='OFNS_DESC', values='COUNT')
#pivot_overall = overall.set_index('HADEVELOPT')
pivot_overall.plot(kind='bar',stacked=True,)
#plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8), title='KY_CD')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 0.8))
plt.title("Top 10 Parks in NYC with Lowest Number of Crimes")
plt.xlabel('PARKS_NM')
plt.ylabel('COUNT')
plt.savefig("overall-crime-park-bottom.png",bbox_inches='tight')
```
