```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import *
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')

# columns required
fields = ['CMPLNT_FR_DT', 'OFNS_DESC']
```

```python
#halloween

#load csv
cleaned = pd.read_csv("cleaned.csv", usecols=fields,)
#dates
dates=['10/31/2006','10/31/2007','10/31/2008', '10/31/2009','10/31/2010','10/31/2011','10/31/2012','10/31/2013', '10/31/2014', '10/31/2015','10/31/2016']
cleaned=cleaned[(cleaned['CMPLNT_FR_DT'].isin(dates))] #pandas interprets NA by default as nulll
cleaned['CMPLNT_FR_DT'] = pd.to_datetime(cleaned['CMPLNT_FR_DT'], format='%m/%d/%Y').apply(lambda x:x.strftime('%m/%d/%Y'))
cleaned = cleaned.groupby(["CMPLNT_FR_DT"]).count()
```

```python
#new year
newyear = pd.read_csv("cleaned.csv", usecols=fields,)
dates=['01/01/2006','01/01/2007','01/01/2008', '01/01/2009','01/01/2001','01/01/2011','01/01/2012','01/01/2013', '01/01/2014', '01/01/2015','01/01/2016']
newyear=newyear[(newyear['CMPLNT_FR_DT'].isin(dates))] #pandas interprets NA by default as nulll
newyear['CMPLNT_FR_DT'] = pd.to_datetime(newyear['CMPLNT_FR_DT'], format='%m/%d/%Y').apply(lambda x:x.strftime('%m/%d/%Y'))
newyear = newyear.groupby(["CMPLNT_FR_DT"]).count()
fig, axes = plt.subplots(nrows=2, ncols=1)
cleaned.plot(ax=axes[0]); axes[0].set_title('Halloween')
#ax = cleaned.plot(label=['abc','xyz'])
newyear.plot(ax=axes[1]); axes[1].set_title('New Year')
#formating graph
plt.title("Crimes Graph")
plt.xlabel('Dates')
plt.ylabel('COUNT')
#plt.savefig("graph.png",bbox_inches='tight')
```
