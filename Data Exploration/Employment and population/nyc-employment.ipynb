```python
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import os
import pandas as pd
import seaborn as sns
import time
plt.style.use('ggplot')

%matplotlib inline

```

```python
empl_p = pd.read_csv("nyc-employment.csv",names=['Dates','Employment (Per Thousands of Person)'])
#removing headers
empl_p = empl_p.loc[1:]
empl_p
```

```python
empl_p['Dates'] = pd.to_datetime(empl_p['Dates'], errors='coerce')

empl_p['Employment (Per Thousands of Person)'] = empl_p['Employment (Per Thousands of Person)'].astype(float)
empl_p.dtypes
empl_p
```

```python
import bokeh.io, bokeh.plotting, bokeh.models
from bokeh.models import DatetimeTickFormatter
bokeh.io.output_notebook()
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file

p1 = figure(x_axis_type="datetime", title=" NYC Employment - 2006 to 2015")
p1.grid.grid_line_alpha=0.5
p1.xaxis.axis_label = 'Year'
p1.yaxis.axis_label = 'Employment (Per Thousands of Person)'
p1.xaxis[0].ticker.desired_num_ticks = 10 #adjust the number of ticks for time axis

p1.line(empl_p['Dates'], empl_p['Employment (Per Thousands of Person)'], color='#FF0000',line_width = 5)

show(p1)
```
