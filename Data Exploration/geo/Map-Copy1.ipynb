```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime as dt
from IPython.display import display
import folium

%matplotlib inline
plt.style.use('ggplot')
```

```python
#load csv
fields = ['Latitude','Longitude','OFNS_DESC']
#oflist=['PETIT LARCENY','HARRASSMENT 2','ASSAULT 3 & RELATED OFFENSES','CRIMINAL MISCHIEF & RELATED OF','GRAND LARCENY','OFF. AGNST PUB ORD SENSBLTY &','DANGEROUS DRUGS','ROBBERY','BURGLARY','FELONY ASSAULT']
oflist=['PETIT LARCENY']

cleaned = pd.read_csv("cleaned.csv", usecols=fields,)
length = cleaned.count
cleaned=cleaned[cleaned['OFNS_DESC'].isin(oflist)] #ofsc list

subfields = ['Station Latitude', 'Station Longitude']
subway = pd.read_csv("subway.csv", usecols=subfields)
#new = cleaned.merge(subway, left_on=['Latitude'], right_on=['Station Latitude'], how='inner')
# new=subway.join(cleaned.set_index(['Latitude','Longitude']), on=['Station Latitude',])
#print(new)
cleaned=cleaned.head(1000)
subway= subway.head(1000)
# cleaned['COUNT']=cleaned.groupby(['LAT','LON'])['LAT','LON'].transform('count')
# cleaned=cleaned.drop_duplicates()
NYC_COORDINATES=[40.8448, -73.8648]
# create empty map zoomed in on San Francisco
map = folium.Map(location=NYC_COORDINATES, zoom_start=11)
#marker_cluster = MarkerCluster().add_to(map)

# icon = folium.features.CustomIcon(icon_url,
#                                       icon_size=(14, 14))
# add a marker for every record in the filtered data, use a clustered view
subway.apply(lambda row:folium.CircleMarker(location=[row["Station Latitude"], row["Station Longitude"]],\
                                             fill=True, # Set fill to True\
                            fill_color='blue',\
                            color = 'blue',\
                            fill_opacity=0.3,clustered_marker = True).add_to(map),axis = 1)

cleaned.apply(lambda row:folium.Marker(location=[row["Latitude"], row["Longitude"]],\
                                      icon=folium.Icon(color='red',icon='info-sign')).add_to(map),axis = 1)


display(map)
#map.save('map.html')
map.save('map.html')
```
