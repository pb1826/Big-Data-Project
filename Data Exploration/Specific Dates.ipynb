```python
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime as dt
from IPython.display import display
%matplotlib inline
plt.style.use('ggplot')

# columns required
fields = ['CMPLNT_FR_DT']

d={
    'halloween' : ['10/31/2006','10/31/2007','10/31/2008', '10/31/2009','10/31/2010','10/31/2011','10/31/2012','10/31/2013', '10/31/2014', '10/31/2015','10/31/2016'],
    'newyear' : ['01/01/2006','01/01/2007','01/01/2008', '01/01/2009','01/01/2001','01/01/2011','01/01/2012','01/01/2013', '01/01/2014', '01/01/2015','01/01/2016'],
    'thanksgiving': ['11/23/2006','11/22/2007','11/27/2008', '11/26/2009','11/25/2010','11/24/2011','11/22/2012','11/28/2013', '11/27/2014', '11/26/2015','11/24/2016'],
    'nineeleven' : ['09/11/2006','09/11/2007','09/11/2008', '09/11/2009','09/11/2010','09/11/2011','09/11/2012','09/11/2013', '09/11/2014', '09/11/2015','09/11/2016'],
    'julyfour' : ['07/04/2006','07/04/2007','07/04/2008', '07/04/2009','07/04/2010','07/04/2011','07/04/2012','07/04/2013', '07/04/2014', '07/04/2015','07/04/2016'],
    'christmas' : ['12/25/2006','12/25/2007','12/25/2008', '12/25/2009','12/25/2010','12/25/2011','12/25/2012','12/25/2013', '12/25/2014', '12/25/2015','12/25/2016'],
    'memorialday' : ['05/30/2006','05/28/2007','05/26/2008', '05/25/2009','05/31/2010','05/30/2011','05/28/2012','05/27/2013', '05/26/2014', '05/25/2015','05/30/2016']

}
dates = pd.DataFrame(d)

dates.to_csv('dates.csv')

```

```python
#halloween

#load csv
cleaned = pd.read_csv("cleaned.csv", usecols=fields,)
#dates
dates['halloween'] = pd.to_datetime(dates['halloween'], format='%m/%d/%Y').apply(lambda x:x.strftime('%m/%d/%Y'))
cleaned=cleaned[(cleaned['CMPLNT_FR_DT'].isin(dates['halloween']))] #pandas interprets NA by default as nulll
cleaned['CMPLNT_FR_DT'] = pd.to_datetime(cleaned['CMPLNT_FR_DT'], format='%m/%d/%Y').apply(lambda x:x.strftime('%m/%d/%Y'))
#cleaned = cleaned.groupby(["CMPLNT_FR_DT"]).count()
cleaned['COUNT']=cleaned.groupby(["CMPLNT_FR_DT"])["CMPLNT_FR_DT"].transform('count')
cleaned=cleaned.drop_duplicates()
# print(cleaned)
# cleaned = cleaned.pivot(index='CMPLNT_FR_DT', columns='OFNS_DESC', values='COUNT')
# cleaned.set_index('CMPLNT_FR_DT')
cleaned['CMPLNT_FR_DT']=pd.to_datetime(cleaned['CMPLNT_FR_DT'])
cleaned.plot(x='CMPLNT_FR_DT',y='COUNT'); 
plt.title("Crimes on Halloween")
plt.xlabel('Dates')
plt.ylabel('COUNT')
plt.savefig("halloween.png",bbox_inches='tight')
```
